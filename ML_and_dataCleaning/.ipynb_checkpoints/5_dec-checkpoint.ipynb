```python
df['Profit'] = df['Profit'].replace({'\$': '', ',': ''}, regex=True)
df['Sales'] = df['Sales'].replace({'\$': '', ',': ''}, regex=True)

# Step 2: Convert the 'Profit' and 'Sales' columns to numeric (float64)
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')

# Step 3: Handle any NaN values if necessary (for example, replace with 0)
df['Profit'] = df['Profit'].fillna(0)
df['Sales'] = df['Sales'].fillna(0)

df.describe()
```

```python
data_summary = round(df.describe().T , 2)
```

```python
data_summary
```

```python
data_summary['cov'] = round(data_summary['std']/data_summary['mean'],2)
```

```python
data_summary
```

```python
numeric_category = []
categorical_category= []

df.drop(columns = ['Row ID','Postal Code'] ,errors='ignore',inplace=True);

for column in df.columns:
    if(df[column].dtype == 'float64' or  df[column].dtype == 'int64'):
        numeric_category.append(column)
    else:
        categorical_category.append(column)
        
print(numeric_category)
print(categorical_category)
```
