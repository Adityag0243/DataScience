```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = np.percentile(col,[25,75])
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5* IQR
    lower_bound = Q1 - 1.5* IQR

    return lower_bound,upper_bound

lb,ub = remove_outlier(df['glucose'])

print(lb)
print(ub)
```

```python

# Function to calculate lower and upper bounds
def remove_outlier(col):
    Q1, Q3 = np.percentile(col.dropna(), [25, 75])  # Exclude NaN values
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    return lower_bound, upper_bound

# Loop through numeric columns
num = df.select_dtypes(include=['float64', 'int64']).columns  # Ensure only numeric columns are selected
for col in num:
    lb, ub = remove_outlier(df[col])
    df[col] = np.where(df[col] > ub, ub, df[col])  # Cap values above upper bound
    df[col] = np.where(df[col] < lb, lb, df[col])  # Cap values below lower bound
```

```python

fig,axes = plt.subplots(nrows = 5 , ncols = 3,figsize=(20,15));

l = num
a = []

palette = sns.color_palette('pastel', n_colors=len(l)) 

for i in range(len(l)):
    row = i//3
    col = i%3
    a.append(sns.boxplot(data = df , x =l[i] ,orient = 'h',ax = axes[row,col], color=palette[i] ,hue=None, legend=False))
    axes[row, col].set_title(f"Boxplot of {l[i]}", fontsize=16)  # Set subplot title

plt.tight_layout();
plt.show();
```
