# Null value Treatment

##### we go for meadian for numerical columns
##### we go for mode for categorical column  (we can do it with KNN imputer)

```python
# Fill NaN values with the median of each column of num

for i in num:
    df[i] = df[i].fillna(df[i].median())
```

```python
df[num].info()
```

```python
df[cat].isnull().sum()
```

```python
# Fill NaN values with the mode of each column of num

for i in cat:
    df[i] = df[i].fillna(df[i].mode()[0])   #[0] bcz there may happen many mode value
```
