```python
import pandas as pd
employees_df = pd.DataFrame({
    'EmpID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie']
})

salaries_df = pd.DataFrame({
    'EmployeeRef': [102, 103, 104],
    'Salary': [70000, 80000, 90000]
})


# our matching columns are empid and employeeref but their name(label) are different so we need to tell by 
# left_on and right_on


joined_df = pd.merge(employees_df, salaries_df, left_on='EmpID', right_on='EmployeeRef')

joined_df
```

```python
import pandas as pd
employees_df = pd.DataFrame({
    'EmpID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie']
})

salaries_df = pd.DataFrame({
    'EmployeeRef': [102, 103, 104],
    'Salary': [70000, 80000, 90000]
})


# our matching columns are empid and employeeref but their name(label) are different so we need to tell by 
# left_on and right_on


joined_df = pd.merge(employees_df, salaries_df, left_on='EmpID', right_on='EmployeeRef',how="inner")

joined_df
```

```python
import pandas as pd
employees_df = pd.DataFrame({
    'EmpID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie']
})

salaries_df = pd.DataFrame({
    'EmployeeRef': [102, 103, 104],
    'Salary': [70000, 80000, 90000]
})


# our matching columns are empid and employeeref but their name(label) are different so we need to tell by 
# left_on and right_on


joined_df = pd.merge(employees_df, salaries_df, left_on='EmpID', right_on='EmployeeRef',how="left")


#here i use how = left which means it will take all data from left(above dataframe) and
# common part from right


joined_df
```

```python
import pandas as pd
employees_df = pd.DataFrame({
    'EmpID': [101, 102, 103],
    'Name': ['Alice', 'Bob', 'Charlie']
})

salaries_df = pd.DataFrame({
    'EmployeeRef': [102, 103, 104],
    'Salary': [70000, 80000, 90000]
})


# our matching columns are empid and employeeref but their name(label) are different so we need to tell by 
# left_on and right_on


joined_df = pd.merge(employees_df, salaries_df, left_on='EmpID', right_on='EmployeeRef',how="right")


#here i use how = left which means it will take all data from left(above dataframe) and
# common part from right


joined_df
```
